```python
import re
from collections import defaultdict

class InvertedIndex:
    def __init__(self):
        self.index = defaultdict(list)

    def tokenize(self, text):
    
        return re.findall(r'\w+', text.lower())

    def add_document(self, doc_id, text):
        words = self.tokenize(text)
        unique_words = set(words)  # Remove duplicates

        for word in unique_words:
            self.index[word].append(doc_id)

    def build_index(self, documents):
        
        for doc_id, text in documents.items():
            self.add_document(doc_id, text)

    def search(self, query):

        words = self.tokenize(query)
        result_sets = []

        for word in words:
            if word in self.index:
                result_sets.append(set(self.index[word]))
            else:
                result_sets.append(set())

        if not result_sets:
            return []

        result = set.intersection(*result_sets)
        return list(result)

    def display(self):
 
        for word in sorted(self.index):
            print(f"{word}: {self.index[word]}")


if __name__ == "__main__":
    documents = {
        1: "This is a cat",
        2: "This is a dog",
        3: "Cat and dog are friends",
        4: "Dogs and cats can live together"
    }

    index = InvertedIndex()
    index.build_index(documents)

    print("----- Inverted Index -----")
    index.display()

    print("\n----- Search Results -----")
    print("Query: 'cat' ->", index.search("cat"))
    print("Query: 'dog' ->", index.search("dog"))
    print("Query: 'cat dog' ->", index.search("cat dog"))
```
